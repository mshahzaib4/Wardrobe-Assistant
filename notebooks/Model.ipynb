```python
# Clean 'discount' column: remove '%' and convert to float
df["discount"] = df["discount"].str.replace('%', '', regex=False).astype(float)

# (Optional) Handle price_pkr if it's not numeric
df["price_pkr"] = pd.to_numeric(df["price_pkr"], errors='coerce')
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib
```

```python
# Define all columns
all_columns = [
    "product_name", "main_category", "Category Style", "categorize_outfit",
    "occasion", "season", "Gender", "fit_type", "fabric", "work_details",
    "shoe_color", "shoe_style", "price_pkr", "discount","color_category",
    "Image URL-src", "Products-href", "web-scraper-start-url"
]

# Columns to drop (non-feature columns or metadata)
drop_cols = [
    "Image URL-src", "Products-href", "web-scraper-start-url",
    "shoe_color", "shoe_style", "discount", "price_pkr", "product_name", "work_details", "color_category"
]

target_col = "categorize_outfit"

# Feature columns (categorical only)
feature_cols = [col for col in all_columns if col not in drop_cols and col != target_col]

# Categorical columns for encoding
categorical_cols = feature_cols.copy()  # Since all features are categorical here

# Drop rows with missing feature data
df = df.dropna(subset=feature_cols)

# Input features and metadata
X = df[feature_cols]
df_meta = df[[target_col, "product_name"]]

# Define preprocessor with only OneHotEncoder (no numerical columns)
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ]
)

# Define KNN model
knn_model = NearestNeighbors(n_neighbors=5, metric='minkowski', p=2, n_jobs=-1)

# Pipeline with preprocessing + KNN
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("knn", knn_model)
])

# Fit pipeline
pipeline.fit(X)
```

```python
joblib.dump(pipeline, "../models/knn_wardrobe_pipeline.pkl")

print("✅ Model trained and saved as knn_wardrobe_pipeline.pkl")
```
